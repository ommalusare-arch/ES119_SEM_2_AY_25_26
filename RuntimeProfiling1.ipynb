Measuring runtime over multiple phases.

```python
# Preparing a list of 10 random numbers
import random
n=10 #input length
X=[]
for i in range(n):
  X.append(random.randint(0,1000))


print(X)
```

```python
# Normalized over multiple runs of the same length
n=10
run=0
for samp in range(100):
  X=[]
  for i in range(n):
    X.append(random.randint(0,1000))
  start = time.process_time()
  Isort(X)
  end = time.process_time()
  run=run+end-start
print(run/100)

```

```python
#Runtime profile over multiple lengths
nval=[]
rval=[]
for n in range(10,1000,10):
  for samp in range(100):
    X=[]
    run =0
    for i in range(n):
      X.append(random.randint(0,1000))
    start = time.process_time()
    Isort(X)
    end = time.process_time()
    run=run+end-start
  nval.append(n)
  rval.append(run/100)
print(nval,"\n")
print(rval)

```

```python
import matplotlib.pyplot as plt

plt.plot(nval, rval)

# labels
plt.xlabel("n")
plt.ylabel("runtime")

# show plot
plt.show()
```
